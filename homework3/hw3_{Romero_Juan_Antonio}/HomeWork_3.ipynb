# Provide an explanation of wich OWASP Top 10 was exploited

The challenges I solved highlighted several key vulnerabilities from the OWASP Top 10. One exercise exploited SQL Injection, where the lack of input validation allowed arbitrary SQL commands to run, exposing sensitive database information and showcasing the risks of poorly implemented query handling. In other tasks, simply inspecting the source code using Ctrl + U revealed sensitive information, such as hidden API keys and administrative links, pointing to Insecure Design. This reflects a failure to consider how attackers might misuse exposed data, emphasizing the importance of securely structuring and hiding sensitive elements. Another issue observed was Insecure Deserialization, where user-manipulated data was not properly validated, allowing tampering that could lead to unauthorized access or execution of arbitrary code. Finally, these vulnerabilities were compounded by Broken Access Control, as I was able to bypass intended restrictions due to insufficient enforcement of user permissions. These examples underscore the need for robust input validation, secure design practices, strict access control mechanisms, and proper handling of serialized data to mitigate such risks effectively.




# What would be a way to patch the vulnerability?

To patch these vulnerabilities effectively, several measures should be implemented. For SQL Injection, using parameterized queries or prepared statements ensures that user input is treated as data, not executable code. Additionally, input validation and sanitization should be applied to all user inputs to prevent malicious entries. For issues related to Insecure Design, sensitive data such as API keys or administrative links should never be hardcoded into the front end or exposed in the source code; instead, they should be stored securely on the server side and accessed through secure APIs. To mitigate Insecure Deserialization, applications should validate and sanitize all serialized data, ensuring it originates from trusted sources, and consider using encryption or signing mechanisms for serialized objects. Finally, for Broken Access Control, strong access control policies should be enforced server-side, using role-based permissions to ensure users can only access data or functionalities they are authorized for. Regular security audits, penetration testing, and secure development practices are crucial to maintaining a robust defense against these vulnerabilities.Also some firewalls to protect the web pages diferent attacks might be a good resource to keep things safe.
