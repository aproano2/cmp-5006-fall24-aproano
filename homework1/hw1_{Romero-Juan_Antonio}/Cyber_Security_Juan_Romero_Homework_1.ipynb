## 1.Prove that if a cryptosystem has perfect secrecy and |K|= |C|= |P|, then every ciphertext is equally probable.

In a cryptosystem with perfect secrecy, the ciphertext reveals no information about the plaintext. This means that knowing the ciphertext does not reduce the uncertainty about the plaintext.

Given:

∣𝐾∣=∣𝐶∣=∣𝑃∣ ∣K∣=∣C∣=∣P∣, where 𝐾 K is the number of keys, 𝐶 C is the number of ciphertexts, and 𝑃 P is the number of plaintexts. Proof:

Perfect Secrecy implies that 𝑃 ( 𝑃 = 𝑝 ∣ 𝐶 = 𝑐 ) = 𝑃 ( 𝑃 = 𝑝 ) P(P=p∣C=c)=P(P=p). This means that the probability of any plaintext 𝑝 p given the ciphertext 𝑐 c is the same as the probability of 𝑝 p without knowing 𝑐 c. In other words, the ciphertext 𝑐 c provides no information about the plaintext 𝑝 p.

Since ∣ 𝐾 ∣ = ∣ 𝐶 ∣ = ∣ 𝑃 ∣ ∣K∣=∣C∣=∣P∣, each key 𝑘 k maps one plaintext to one ciphertext. This is a one-to-one mapping (bijective function).

Because of this bijection and the equiprobability of the keys, the probability of each ciphertext is equal. For each ciphertext 𝑐 c, there is exactly one key 𝑘 k and one plaintext 𝑝 p that could produce it.

A substitution cipher encodes a message by replacing each letter or symbol in the plaintext with another letter or symbol according to a fixed system. This method can range from simple single-letter replacements to more complex systems involving multiple substitutions.Below we are going to use a code to try and decypher this messsage by using substitution cypher.

```python
from collections import Counter

def frequency_analysis(ciphertext):
    """Analyze the frequency of letters in the ciphertext."""
    # Remove non-letter characters and convert to uppercase
    ciphertext = ''.join(filter(str.isalpha, ciphertext)).upper()

    # Count the frequency of each letter
    freq = Counter(ciphertext)

    # Sort letters by frequency
    total_letters = sum(freq.values())
    sorted_freq = sorted(freq.items(), key=lambda item: item[1], reverse=True)

    # Calculate frequency percentage
    frequency_percentage = {letter: count / total_letters for letter, count in sorted_freq}
    return frequency_percentage

def suggest_substitutions(frequency_percentage):
    """Suggest substitutions based on the frequency analysis."""
    # Typical letter frequencies in English (percentage)
    english_freq_order = 'ETAOINSHRDLCUMWFGYPBVKJXQZ'  # Most common letters in order

    # Suggest substitution based on frequency
    substitutions = {}
    sorted_letters = [letter for letter, _ in frequency_percentage.items()]

    for i, letter in enumerate(sorted_letters):
        if i < len(english_freq_order):
            substitutions[letter] = english_freq_order[i]

    return substitutions

def decode_substitution(ciphertext, substitutions):
    """Decode the ciphertext using the provided substitutions."""
    decoded_message = []

    for char in ciphertext:
        if char.upper() in substitutions:
            decoded_char = substitutions[char.upper()]
            # Preserve the case
            if char.islower():
                decoded_message.append(decoded_char.lower())
            else:
                decoded_message.append(decoded_char)
        else:
            decoded_message.append(char)  # Non-alphabetic characters remain unchanged

    return ''.join(decoded_message)

def main():
    ciphertext = "EMGLOSUDCGDNCUSWYSFHNSFCYKDPUMLWGYICOXYSIPJCKQPKUGKMGOLICGINCGACKSNISACYKZSCKXECJCKSHYSXCGOIDPKZCNKSHICGIWYGKKGKGOLDSILKGOIUSIGLEDSPWZUGFZCCNDGYYSFUSZCNXEOJNCGYEOWEUPXEZGACGNFGLKNSACIGOIYCKXCJUCIUZCFZCCNDGYYSFEUEKUZCSOCFZCCNCIACZEJNCSHFZEJZEGMXCYHCJUMGKUCY"

    # Step 1: Analyze frequency of letters in the ciphertext
    freq_analysis = frequency_analysis(ciphertext)
    print("Frequency Analysis:")
    for letter, freq in freq_analysis.items():
        print(f"{letter}: {freq:.4f}")

    # Step 2: Suggest substitutions based on frequency analysis
    substitutions = suggest_substitutions(freq_analysis)
    print("\nSuggested Substitutions:")
    for letter, suggestion in substitutions.items():
        print(f"{letter} -> {suggestion}")

    # Step 3: Decode the message
    decoded_message = decode_substitution(ciphertext, substitutions)
    print("\nDecoded Message:")
    print(decoded_message)

if __name__ == "__main__":
    main()
```

A Vigenère cipher is a method of encrypting alphabetic text by using a simple form of polyalphabetic substitution, where a keyword determines the shift of each letter in the plaintext. Each letter of the keyword corresponds to a letter in the plaintext, creating a varying shift for each letter based on the position of the corresponding letter in the keyword.

```python
import string
from collections import Counter

# Function to repeat the key to match the length of the text
def repeat_key(key, length):
    return (key * (length // len(key)) + key[:length % len(key)]).upper()

# Function to calculate the Index of Coincidence for a given string
def index_of_coincidence(text):
    N = len(text)
    freqs = Counter(text)
    return sum([f * (f - 1) for f in freqs.values()]) / (N * (N - 1)) if N > 1 else 0

# Function to guess the key length by splitting the text and calculating the IC for each
def guess_key_length(ciphertext, max_key_len=20):
    ciphertext = ''.join([c.upper() for c in ciphertext if c.isalpha()])
    avg_ic = []

    for key_len in range(1, max_key_len + 1):
        ic_sum = 0
        for i in range(key_len):
            # Get every i-th letter (like splitting into columns)
            nth_letters = ciphertext[i::key_len]
            ic_sum += index_of_coincidence(nth_letters)
        avg_ic.append((key_len, ic_sum / key_len))

    # Return key lengths with high IC values
    return sorted(avg_ic, key=lambda x: -x[1])

# Function to split ciphertext into columns based on the guessed key length
def split_into_columns(ciphertext, key_len):
    ciphertext = ''.join([c.upper() for c in ciphertext if c.isalpha()])
    columns = ['' for _ in range(key_len)]
    for i, char in enumerate(ciphertext):
        columns[i % key_len] += char
    return columns

# Frequency analysis of a text column
def frequency_analysis(text):
    text = text.upper()
    text = ''.join([c for c in text if c.isalpha()])
    letter_freq = Counter(text)
    total_letters = sum(letter_freq.values())

    # Print frequencies of each letter
    for letter in string.ascii_uppercase:
        freq = letter_freq[letter] / total_letters if total_letters > 0 else 0
        print(f'{letter}: {freq:.4f}')

# Function to decrypt using Vigenère Cipher
def vigenere_decrypt(ciphertext, key):
    alphabet = string.ascii_uppercase
    plaintext = ""

    key = repeat_key(key, len(ciphertext))

    for i in range(len(ciphertext)):
        if ciphertext[i].isalpha():
            shift = ord(key[i]) - ord('A')
            decrypted_char = chr((ord(ciphertext[i].upper()) - ord('A') - shift + 26) % 26 + ord('A'))
            plaintext += decrypted_char
        else:
            plaintext += ciphertext[i]

    return plaintext

# Example ciphertext (replace with your own ciphertext)
ciphertext = "KCCPKBGUFDPHQTYAVINRRTMVGRKDNBVFDETDGILTXRGUDDKOTFMBPVGEGLTGCKQRACQCWDNAWCRXIZAKFTLEWRPTYCQKYVXCHKFTPONCQQRHJVAJUWETMCMSPKQDYHJVDAHCTRLSVSKCGCZQQDZXGSFRLSWCWSJTBHAFSIASPRJAHKJRJUMVGKMITZHFPDISPZLVLGWTFPLKKEBDPGCEBSHCTJRWXBAFSPEZQNRWXCVYCGAONWDDKACKAWBBIKFTIOVKCGGHJVLNHIFFSQESVYCLACNVRWBBIREPBBVFEXOSCDYGZWPFDTKFQIYCWHJVLNHIQIBTKHJVNPIST"

# Step 1: Guess the key length using the Index of Coincidence
key_length_guesses = guess_key_length(ciphertext, max_key_len=10)
print("Possible key lengths based on Index of Coincidence:", key_length_guesses)

# Assuming the first key length guess is correct, let's perform frequency analysis on each column
key_len = key_length_guesses[0][0]  # Choose the most probable key length
columns = split_into_columns(ciphertext, key_len)

# Step 2: Perform frequency analysis for each column
for i, column in enumerate(columns):
    print(f"\nFrequency analysis for column {i + 1}:")
    frequency_analysis(column)

# Step 3: guess the keyword
key = "CRYPTO"

# Step 4: Decrypt the ciphertext using the guessed key
decrypted_text = vigenere_decrypt(ciphertext, key)
print(f"\nDecrypted text: {decrypted_text}")
```
