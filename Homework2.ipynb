##### **1.1. What are cryptographic hash functions?**

A cryptographic hash function is an algorithm that takes any amount of data like a password and produces a fixed-size, encrypted output called a hash. This hash can be stored in place of the actual password and later used to verify the user’s identity.

Several key features of cryptographic hash functions are essential for secure password storage.

##### **1.2. What are the main properties of cryptographic hash functions?**

The main properties of crytographic hash functions are the following:

- **One-way function (non-reversibility):** A good hash function makes it nearly impossible to reverse-engineer the original password from the hash.
- **Avalanche effect (diffusion):** A small change in the original password should produce a drastically different hash. Even a one-bit change should alter about half the bits in the hash, making the output unpredictable.
- **Consistency (determinism):** The same password should always generate the same hash value.
- **Collision resistance:** It should be very difficult to find two different passwords that produce the same hash.
- **Unpredictability:** The hash value shouldn’t be predictable based on the password itself.

These properties help ensure that password storage is as secure as possible.

##### **1.3. Pre-image resistance**

Preimage resistance is a key property of hash functions, ensuring that it is computationally infeasible to find the original input just by knowing the output hash.

For a hash function to be considered preimage-resistant, it should generate a hash output of at least 80 bits. Preimage resistance is distinct from other security properties like second preimage resistance and collision resistance. For a hash function to be "one-way," it must first satisfy both preimage and second preimage resistance.

### **First Definition**
In the first approach, preimage resistance means that, given a hash output, it should be extremely difficult to find the original input that produced it. A secure hash function should be uniform, meaning each input has an equal chance of producing any output. If an attacker could identify the input by guessing, they would break preimage resistance. However, by selecting the hash function at random, it becomes exceedingly challenging for the attacker to guess correctly.

### **Second Definition**
In the second approach, preimage resistance involves selecting from multiple possible hash functions. Once a specific hash function is chosen, a random message is hashed, and the result is given to an attacker who must find the input that created it. If the attacker can do so, preimage resistance is compromised. This definition is particularly practical, as it models hash functions used in real-world applications.

### **Third Definition**
In the third approach, a random hash function and a random input (of a specified length) are chosen. The hash of this input is then provided to the attacker along with details about the chosen hash function. The attacker’s task is to identify the original input from the hash. If they succeed, preimage resistance has been broken.


##### **1.4. Second pre-image resistance**

Second preimage resistance, also known as “weak collision resistance,” is a property of hash functions that makes it extremely difficult to find a second, different input that produces the same hash as a given input. In other words, if you already have one input and its hash, finding a different input with the same hash should be nearly impossible.

This property shares similarities with preimage resistance, as both are types of one-way hash functions that prevent reverse calculation. However, second preimage resistance specifically applies to functions with similar input and output domains.

To be second preimage resistant, a hash function typically needs a minimum length of 90 bits. Sometimes, people confuse second preimage resistance with preimage resistance due to their similarities. However, preimage resistance is foundational: for a hash function to be fully second preimage resistant, it must first satisfy the requirements of preimage resistance.


##### **1.5. Collision resistance**

Collision resistance is another essential property of cryptographic hash functions, closely related to second preimage resistance. Because of their similarities, collision resistance is sometimes called “weak collision resistance.” However, for a hash function to be considered truly collision-resistant, it generally requires a minimum output length of 160 bits.

This property means that it should be extremely difficult to find two different inputs that produce the same hash output. Although many hash functions are designed to be collision-resistant, some have been found vulnerable over time. For example, MD5 and SHA-1, once widely used, were eventually proven susceptible to attacks that could find collisions more efficiently than brute force alone. To ensure collision resistance, modern cryptographic hash functions are often designed to be as hard to break as solving complex mathematical problems like integer factorization or discrete logarithms these functions are known as “provably secure.”

In summary, preimage resistance, second preimage resistance, and collision resistance are all critical properties of secure hash functions. Preimage resistance is foundational, as a hash function must be preimage resistant before it can achieve second preimage resistance. Collision resistance is also closely related to second preimage resistance, which is why both are sometimes referred to as “weak collision resistance.”

##### **1.6. How do these properties contribute to security?**

The security properties of cryptographic hash functions pre-image resistance, second pre-image resistance, and collision resistance—are essential for safeguarding digital information:

1. **Pre-image Resistance**: This property ensures that, given a hash output, it’s computationally infeasible to reverse-engineer the original input. This prevents attackers from retrieving sensitive data, making it ideal for securing passwords and other confidential information.

2. **Second Pre-image Resistance**: This prevents attackers from finding a different input that produces the same hash as a given message. It ensures data authenticity, as seen in digital signatures, where it blocks attempts to substitute genuine documents with fake ones.

3. **Collision Resistance**: By making it infeasible to find two distinct inputs with the same hash, this property protects against collision attacks. This is vital for blockchains and digital certificates, where unique data identification is critical for system integrity.

Together, these properties make hash functions trustworthy for securing data integrity, ensuring authenticity, and protecting sensitive information in applications across digital systems.

##### **2.1. What are the most commonly used hash functions?**

**MD5 (Message Digest Algorithm 5):** Developed by Ronald Rivest in 1991, MD5 became one of the most widely used hash functions for checksums, digital signatures, and integrity checks. MD5 outputs a 128-bit hash value and is known for its speed, which led to its extensive use. However, MD5’s vulnerabilities, especially susceptibility to collision attacks (where two different inputs produce the same hash), have rendered it insecure for cryptographic use.

**SHA-1 (Secure Hash Algorithm 1):** Released by the National Security Agency (NSA) in 1993, SHA-1 outputs a 160-bit hash and became a standard hash function for digital signatures, SSL/TLS certificates, and other security protocols. Over time, however, SHA-1 became vulnerable to collision attacks due to advancements in computational power and cryptographic techniques. In 2017, the "SHAttered" attack demonstrated the first practical collision for SHA-1, leading to widespread deprecation in favor of more secure alternatives.

**SHA-2 Family:** This family of hash functions, also designed by the NSA, includes variations such as SHA-224, SHA-256, SHA-384, and SHA-512, each producing hash values with bit lengths that correspond to their names (e.g., SHA-256 produces a 256-bit output). The SHA-2 family has been the recommended standard for secure hashing since SHA-1's vulnerabilities surfaced. SHA-256, in particular, is heavily used in blockchain applications, SSL/TLS certificates, and various encryption protocols.

**SHA-3:** The newest addition to the SHA family, SHA-3 was developed through a public competition held by the National Institute of Standards and Technology (NIST) to find a backup to SHA-2. Based on the Keccak algorithm, SHA-3 is constructed using a different design than SHA-1 and SHA-2, making it theoretically resistant to the vulnerabilities found in those older algorithms. SHA-3 has versions with output sizes of 224, 256, 384, and 512 bits, similar to the SHA-2 family.

##### **2.3. Why have some hash functions been deprecated?**

Hash functions like MD5 and SHA-1 were deprecated due to advances in collision attacks:

Collision Attacks: A hash collision occurs when two different inputs produce the same hash output. In cryptographic applications, this vulnerability can allow attackers to substitute data without detection, severely compromising security. For example, an attacker could create a fraudulent SSL certificate with the same SHA-1 hash as a legitimate one, tricking systems into accepting it as authentic.

Preimage and Second Preimage Attacks: Although collision attacks are the main reason for MD5 and SHA-1’s deprecation, they are also theoretically susceptible to preimage attacks (finding an input that matches a specific hash) and second preimage attacks (finding another input that hashes to the same value as a specific input). Though these attacks are not as practical as collision attacks, they highlight further weaknesses in these older algorithms.

As these vulnerabilities became more practical, particularly with the "SHAttered" collision demonstrated for SHA-1 in 2017, major organizations like NIST, Microsoft, and Google began phasing out MD5 and SHA-1 from their security infrastructures.

Parallel Processing Compatibility: SHA-3 is designed to be more compatible with parallel processing architectures. Though SHA-3 provides an additional layer of cryptographic security, SHA-2 is still widely regarded as secure, and SHA-3 is mainly reserved as a backup in the event of a breakthrough that affects SHA-2’s reliability. ##### **3.1. How are hash functions used in digital signatures?**

**Creating a Hash:** When a sender wants to send a signed document, they first generate a hash of the document using a hash function. This hash is a fixed-size string of characters that represents the document's contents uniquely.

**Signing the Hash:** The sender then encrypts this hash with their private key, creating the digital signature. This ensures that the signature is unique to both the document and the sender.

**Sending the Document:** The sender transmits the document along with the digital signature to the recipient.

**Verifying the Signature:** Upon receiving the document and the signature, the recipient also generates a hash of the received document. They then decrypt the signature using the sender's public key to retrieve the original hash. If the two hashes match, it confirms that the document hasn't been altered and verifies the sender's identity.

##### **3.2. What role do hash functions play in password storage?**

**Hashing the Password:** When a user creates an account, their password is processed through a hash function. This creates a hashed value that represents the password, which is then stored in the database instead of the actual password.

**Salting:** To enhance security, a random value (salt) is often added to the password before hashing. This ensures that even if two users have the same password, their hashes will differ, making it more challenging for attackers to use precomputed tables (rainbow tables) to crack passwords.

**Verification:** When a user attempts to log in, their entered password is hashed again using the same method, and the resulting hash is compared to the stored hash. If they match, access is granted; otherwise, it is denied.

##### **3.3. How are hash functions used in blockchain technology?**

**Block Integrity:** Each block in a blockchain contains a hash of the previous block, linking them together in a chain. This ensures that if any block's data is altered, its hash would change, breaking the chain and alerting users to tampering.

**Transaction Verification:** When transactions are added to the blockchain, they are hashed to create a unique identifier for the transaction. This ensures that transactions are securely linked and can be verified without exposing sensitive data.

**Consensus Mechanisms:** In systems like Bitcoin, hash functions are used in the Proof of Work consensus algorithm, where miners must solve complex mathematical problems involving hashes to add new blocks to the chain. This process helps secure the network and validate transactions.


##### **3.4. What is a HMAC and why is it important**

HMAC stands for Hash-based Message Authentication Code. It is a specific type of message authentication code that uses a cryptographic hash function in combination with a secret key.

**Integrity and Authenticity:** HMACs ensure that a message has not been altered and verifies the authenticity of the sender. By using a secret key, HMACs can confirm that only authorized parties can produce the correct code.

**Resistance to Attacks:** HMACs are designed to be resistant to certain types of cryptographic attacks, including collision attacks. The use of a secret key in conjunction with the hash function enhances security compared to using the hash function alone.

**Widespread Use:** HMACs are widely used in various security protocols, including TLS, IPsec, and JSON Web Tokens (JWT), making them essential for secure communications over the internet.

**1.1. How will keys be generated?**

For key generation, we will rely on encryption systems, using two types in this case.

Once a secure channel for communication is established, we will generate the second key using symmetric encryption (AES) to encrypt the messages shared between them. We chose AES because it performs better in scenarios where large volumes of data need to be encrypted. For AES, we will use a 256-bit key length, which is secure and efficient for messages.


**1.2. How will keys be exchanged?**

To ensure a secure key exchange between Alice and Bob, we will use the Diffie-Hellman (DH) protocol based on Elliptic Curve Cryptography (ECDH). This protocol enables both parties to privately generate a shared key without transmitting sensitive information that could be intercepted by an attacker. During the exchange, Alice and Bob share their public keys, generated through elliptic curve algorithms, and then use their private keys to compute the same shared key.

For example, Alice generates her key pair and sends her public key to Bob. Bob does the same and sends his public key to Alice. Upon receiving it, they both compute a shared key using their private key and the other's public key. The security of ECDH lies in the complex mathematics of elliptic curves, which makes it practically impossible for an attacker to derive the shared key without access to the private keys.

To further secure the exchange and prevent man-in-the-middle attacks, digital certificates will be used to authenticate the public keys of Alice and Bob. These certificates, issued by a trusted Certificate Authority (CA), ensure each public key genuinely belongs to the respective party. This way, ECDH combined with digital certificates creates a secure key exchange, where only Alice and Bob can access the shared key for safe communication.

Encryption keys will be stored securely on each device, with an additional level of encryption applied to protect them against unauthorized access. Besides encryption, it is recommended that keys be stored in specialized hardware such as Hardware Security Modules (HSM) or secure execution environments like Secure Enclave. **3.2. Message integrity checks**

To ensure messages have not been altered, a Message Authentication Code (MAC) generated through AES-GCM will be used, which includes a "tag" to verify message integrity. Upon receiving a message, the receiver recalculates the MAC and compares it with the original. If the MACs match, the message is intact; if not, the user is alerted that the message may have been altered.

**3.3. Replay attack prevention**

To prevent attackers from replaying old messages (replay attacks), each message will include a sequence number or unique nonce. This value changes with each message and is validated by the receiver to ensure there are no duplicates. If a message arrives with a previously recorded nonce or sequence number, it is flagged as a replay attempt and discarded.

**3.4. Perfect forward secrecy (optional)**

To protect the confidentiality of messages in case a private key is compromised, perfect forward secrecy will be implemented by rotating keys in each session. This way, even if an attacker gains access to a private key in the future, they won’t be able to decrypt past sessions since each session uses a unique key derived from the initial ECDH key exchange.
